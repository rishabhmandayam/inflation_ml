```python
tcode5_cols = data_dictionary.filter(pl.col("tcode") == 5)["fred"].to_list()
tcode6_cols = data_dictionary.filter(pl.col("tcode") == 6)["fred"].to_list()
cols_to_transform = tcode5_cols + tcode6_cols
```

```python
def add_transformations(df, columns):
    # Create expressions for all transformations
    mom_exprs = []
    yoy_exprs = []
    
    for col in columns:
        if col in df.columns:
            mom_expr = (
                ((pl.col(col) - pl.col(col).shift(1)) / pl.col(col).shift(1) * 100)
                .alias(f"{col}_MOM")
            )
            mom_exprs.append(mom_expr)
            
            yoy_expr = (
                ((pl.col(col) - pl.col(col).shift(12)) / pl.col(col).shift(12) * 100)
                .alias(f"{col}_YOY")
            )
            yoy_exprs.append(yoy_expr)
    
    return df.with_columns(mom_exprs + yoy_exprs)

df = add_transformations(df, cols_to_transform)

print(f"Added {len(cols_to_transform)} MOM and {len(cols_to_transform)} YOY transformations")
```

```python
numerical_cols = [col for col in df.columns if col not in ['sasdate', 'date', 'CPIAUCSL_MOM']]
correlation_results = []
for col in numerical_cols:
    if df[col].dtype not in [pl.Float64, pl.Int64]:
        continue
        
    temp_df = df.select([
        col,
        'CPIAUCSL_MOM'
    ]).with_columns([
        pl.col(col).shift(-1).alias(f"{col}_lag1")  
    ])
    
    correlation = temp_df.filter(
        pl.col(f"{col}_lag1").is_not_null() & 
        pl.col("CPIAUCSL_MOM").is_not_null()
    ).select(
        pl.corr(f"{col}_lag1", "CPIAUCSL_MOM").alias("correlation")
    )[0, 0]
    
    correlation_results.append({"feature": col, "correlation": correlation})

corr_df = pl.DataFrame(correlation_results)
corr_df = corr_df.with_columns(pl.col("correlation").abs().alias("abs_correlation"))
corr_df = corr_df.sort("abs_correlation", descending=True)

print("Top 15 features with highest correlation to next month's inflation:")
print(corr_df.head(15))
```

## Feature Selection using Lasso

```python
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
import matplotlib.pyplot as plt
```

```python
def prepare_lagged_features(input_df, target_col='CPIAUCSL_MOM', lag=1):    
    available_cols = [col for col in numerical_cols if col in input_df.columns]
    print(available_cols)
    
    lagged_cols = [pl.col(col).shift(lag).alias(f"{col}_lag{lag}") for col in available_cols]
    
    # Apply lagged columns to the full DataFrame
    model_df = input_df.with_columns(lagged_cols)
    # Then select only the target and the new lagged columns
    model_df = model_df.select([pl.col(target_col)] + [pl.col(f"{col}_lag{lag}") for col in available_cols])
    model_df = model_df.drop_nulls()
    
    return model_df
```

```python
model_df = prepare_lagged_features(df)
y = model_df['CPIAUCSL_MOM'].to_numpy()
feature_names = [col for col in model_df.columns if col != 'CPIAUCSL_MOM']
X = model_df.select(feature_names).to_numpy()
```

```python
print(model_df.head(10))
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

tscv = TimeSeriesSplit(n_splits=5)
```

```python
lasso_cv = LassoCV(
    cv=tscv,
    random_state=42,
    max_iter=10000,
    tol=1e-4,
    n_alphas=100,
    alphas=np.logspace(-6, 2, 100)
)

lasso_cv.fit(X_scaled, y)
```

```python
print(y)
```

```python
selected_features = []
selected_coefficients = []

for feature, coef in zip(feature_names, lasso_cv.coef_):
    if abs(coef) > 0:
        selected_features.append(feature)
        selected_coefficients.append(coef)
```

```python
sorted_idx = np.argsort(np.abs(selected_coefficients))[::-1]
sorted_features = [selected_features[i] for i in sorted_idx]
sorted_coefficients = [selected_coefficients[i] for i in sorted_idx]

# Print results
print(f"LASSO selected {len(sorted_features)} out of {len(feature_names)} features")
print(f"Optimal alpha: {lasso_cv.alpha_:.6f}")
```

```python
results = pl.DataFrame({
    'Feature': sorted_features[:20],
    'Coefficient': sorted_coefficients[:20]
})
print("\nTop 20 features by importance:")
print(results)
```

```python
plt.figure(figsize=(10, 6))
alphas = lasso_cv.alphas_
mse = lasso_cv.mse_path_.mean(axis=1)
print(mse)
plt.semilogx(alphas, mse)
plt.axvline(lasso_cv.alpha_, color='k', linestyle='--')
plt.xlabel('Alpha')
plt.ylabel('Mean Squared Error')
plt.title('Cross-Validation Error vs. Alpha')
plt.tight_layout()
plt.show()
```

```python

original_feature_names = [feature.replace('_lag1', '') for feature in sorted_features]

selected_data = df.select([
    'sasdate', 
    'CPIAUCSL_MOM',  
    *original_feature_names  
])

output_path = '../data/selected_features_inflation.csv'
selected_data.write_csv(output_path)

print(f"Saved {len(original_feature_names)} selected features and target variable to {output_path}")
print(f"Selected features: {original_feature_names[:10]}..." if len(original_feature_names) > 10 else original_feature_names)

feature_importance = pl.DataFrame({
    'Feature': sorted_features,
    'Coefficient': sorted_coefficients,
    'AbsCoefficient': [abs(c) for c in sorted_coefficients]
}).sort('AbsCoefficient', descending=True)

feature_importance.write_csv('../data/feature_importance_lasso.csv')
print(f"Saved feature importance information to ../data/feature_importance_lasso.csv")
```
